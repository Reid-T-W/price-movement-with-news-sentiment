```python
# Load headline data
import dask.dataframe as dd
dask_dataframe_analyst = dd.read_csv("../data/raw/raw_analyst_ratings.csv")
```

```python
df_aapl = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'AAPL'].compute()
df_amzn = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'AMZN'].compute()
df_goog = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'GOOG'].compute()
df_meta = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'META'].compute()
df_msft = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'MSFT'].compute()
df_nvda = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'NVDA'].compute()
df_tsla = dask_dataframe_analyst[dask_dataframe_analyst['stock'] == 'TSLA'].compute()
```

```python
df_aapl.head()
```

```python
rating_dataframes = [["aapl", df_aapl], ["amzn", df_amzn], ["goog", df_goog], ["meta", df_meta], ["msft", df_msft], ["nvda", df_nvda], ["tsla", df_tsla]]
```

```python
def analyze_ratings_perstock(df):
    print("--------------------------------")
    print(f"No of records for Ticker {df[0]}")
    print(df[1].shape)
```

```python
for df in rating_dataframes:
    analyze_ratings_perstock(df)
```

```python
from textblob import TextBlob
# Extract sentiments
def calculate_sentiment(text):
    return TextBlob(text).sentiment.polarity

def bulk_sentiment_calcuation(df):
    # Applying the sentiment analysis function to your headlines
    df[1]['sentiment'] = df[1]['headline'].apply(calculate_sentiment)
    print("---------------------------------")
    print(f"Ticker {df[0]} sample sentiments")
    print(df[1][["headline", "sentiment"]].head())

for df in rating_dataframes:
    bulk_sentiment_calcuation(df)
```

```python
# Drop unnecessary rows
def drop_unnecessary_cols(df):
    df[1] = df[1].drop("url", axis=1)
    print("---------------------------------")
    print(f"Ticker {df[0]} columns")
    print(df[1].columns)

for df in rating_dataframes:
    drop_unnecessary_cols(df)
```
