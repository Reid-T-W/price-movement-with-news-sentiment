```python
import pandas as pd
df_aapl = pd.read_csv(r'../data/raw/yfinance_data/AAPL_historical_data.csv')
df_amzn = pd.read_csv(r'../data/raw/yfinance_data/AMZN_historical_data.csv')
df_goog = pd.read_csv(r'../data/raw/yfinance_data/GOOG_historical_data.csv')
df_meta = pd.read_csv(r'../data/raw/yfinance_data/META_historical_data.csv')
df_msft = pd.read_csv(r'../data/raw/yfinance_data/MSFT_historical_data.csv')
df_nvda = pd.read_csv(r'../data/raw/yfinance_data/NVDA_historical_data.csv')
df_tsla = pd.read_csv(r'../data/raw/yfinance_data/TSLA_historical_data.csv')

financial_dataframes = [["aapl", df_aapl], ["amzn", df_amzn], ["goog", df_goog], ["meta", df_meta], ["msft", df_msft], ["nvda", df_nvda], ["tsla", df_tsla]]
```

```python
def describe_data(df):
    print("------------------------------------")
    print(f"TICKER {df[0]}")
    print(df[1].describe())
```

```python
for df in financial_dataframes:
    describe_data(df)
```

```python
import talib
import mplfinance as mpf

def plot_sma(df):
    # Prepare your DataFrame
    df[1]['Date'] = pd.to_datetime(df[1]['Date'])
    df[1].set_index('Date', inplace=True)

    # Calculate the 5-period SMA
    df['SMA_5'] = talib.SMA(df[1]['Close'], timeperiod=5)

    # Plot candlestick chart with SMA overlay
    mpf.plot(
        df[1],
        type='candle',
        mav=(5,),         # Moving average window(s)
        volume=True,      # Show volume subplot
        title=f'{df[0].upper} Candlestick with 5-period SMA'
    )
    # sma = talib.SMA(df.Close,5)
    # candlestick_ochl(df)
    # plt.plot(sma, 'k')
```

```python
for df in financial_dataframes:
    plot_sma(df)
```

```python
import matplotlib.pyplot as plt
import talib
```

```python
from matplotlib.pyplot import spy


def plot_rsi(df):
    print("------------------------------------")    
    plt.figure(figsize=[10,7])
    rsi = talib.RSI(df[1].Close, 14)
    plt.plot(rsi)
    plt.title(f"RSI for {df[0].upper()}")
    plt.xlabel("Index")
    plt.ylabel("RSI")
    plt.show()
```

```python
for df in financial_dataframes:
    plot_rsi(df)
```
